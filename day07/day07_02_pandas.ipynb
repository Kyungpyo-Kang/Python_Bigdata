# 데이터 시각화

```python
import pandas as pd
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# seaborn : 좀 더 예쁜 시각하를 위한 외부 패키지
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.sepal_length[:20].plot(kind='bar')
plt.title('iris')
plt.xlabel('data')
plt.ylabel('length')
pass
```

```python
# 그룹분석으로 각 붓꽃의 특징값의 평균 구하기
df2 = iris.groupby(iris.species).mean()
```

```python
df2.plot.bar()
pass
```

## pie 차트
- 전체 중에서 나의 비율을 표시할 때 유용한 차트

```python
# 실습하기
# iris 종의 갯수를 조회하라.
df3 = iris.species.value_counts()
```

```python
df3.plot.pie(autopct='%.2f%%')
plt.title('species_pie')
plt.ylabel('')
plt.show()
pass
```

```python
# 실습하기 0~140번째 데이터만으로 iris2만들기
iris2 = iris[:140]
```

```python
df4 = iris2.species.value_counts()
```

```python
df4.plot.pie(autopct='%.2f%%')
plt.title('species_pie')
plt.ylabel('')
plt.show()
pass
```

```python
# hist : 빈도 수
iris.plot.hist()
# 해석방법
# ex) petal_width는 0~1 구간에 출현하는 데이터 갯수가 약 50개이다.
pass
```

```python
tips = sns.load_dataset('tips')
```

#### 산포도 : 데이터가 어떻게 분포되어 있는지를 확인

```python
sns.relplot(x='total_bill', y='tip', data=tips)
pass
```

```python
# 여자와 남자 중에서 팁은 누가 더 많이 낼까?
# 비슷하게 낸다.
sns.relplot(x='sex', y='tip', data=tips)
pass
```

```python
# 남녀의 비율 보기
sns.barplot(x='sex', y='tip', data=tips)
pass
```

```python
# 총 계산 금액의 분포도와 밀도 보기
sns.distplot(tips['total_bill'])
pass
```

```python
# 전체 계산 금액 중에서 팁에 흡연 여부가 관련이 있는가?
# -hue 옵션을 지정하면 카테고리형으로 구분이 가능
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker')
pass
```

```python
# 실습하기
# 테이블의 size에 따른 팁을 시각화
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', size='size')
pass
```

```python
# 점심과 저녁을 카테고리로 나누기
sns.catplot(x='time', y='tip', data=tips)
pass
```

```python
# 실습하기
# 점심과 저녁 중에서 흡연자 여부를 포함하여 시각화
sns.catplot(x='time', y='tip', data=tips, hue='smoker')
pass
```

```python
# 팁은 주중(목,금)과 주말 중에서 언제 더 많이 낼까?
sns.catplot(x='day', y='tip', data=tips)
pass
```

```python
# 형태를 가진 그래프
sns.swarmplot(x='day', y='total_bill', data=tips)
pass
```
