# treemap 예제
- 각 카테고리별로 특정 데이터의 집합이 전체 데이터에서 차지하는 비율을 면적으로 표시하는 plot

```python
import matplotlib
import matplotlib.pyplot as plt
import squarify
import pandas as pd
```

```python
sizes = [1,2,10,20]
label = ['groupA','groupB','groupC','groupD']
```

```python
squarify.plot(sizes=sizes
              , label=label
              , alpha=0.8
             , color=['red','green','yellow','gray'])
pass
```

```python
# 실습하기
# banana 데이터를 df 변수에 저장
# -header=None이 오류나면 header=False
df = pd.read_csv('https://raw.githubusercontent.com/luxdolorosa/data_set/master/banana/banana.csv', header=None)
```

```python
# df의 columns를 country, ratio로 변경하기
df.columns = ['country','ratio']
df
```

```python
# 실습하기
# df의 데이터를 이용하여 treemap 그리기
# treemap size = ratio
# label = country
```

```python
squarify.plot(sizes=df.ratio
              , label=df.country
              , alpha=0.8
              , color=['pink','orange','yellow','green','lightblue','gray'])
pass
```

```python
df['label'] = df.loc[:,'country']+'('+df.loc[:,'ratio'].astype('str')+'%)'
squarify.plot(sizes=df.ratio
              , label=df.label
              , alpha=0.8
              , color=['pink','orange','yellow','green','lightblue','gray'])
pass
```
